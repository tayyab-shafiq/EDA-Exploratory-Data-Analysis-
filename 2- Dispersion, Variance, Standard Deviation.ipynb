```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
import seaborn as sns

import seaborn as sns
sns.set(color_codes=True)
sns.set_style("white")     # See more styling options here: https://seaborn.pydata.org/tutorial/aesthetics.html
```

```python
sample_size = 5000
uniform_val_data = np.random.uniform(low=1.0, high=30.0, size=sample_size)
```

```python
print(uniform_val_data)
```

```python
sns.distplot(uniform_val_data, bins=20, kde=False)
plt.title("Uniform Values")
plt.show()
```

```python
normal_val_data = np.random.normal(loc=15, scale=10, size=sample_size)
# loc = zyada tar data 15 ke qareeb ho & slightly barhta jaye  
```

```python
sns.distplot(normal_val_data, bins=20, kde=False)
plt.title("Normal Values")
plt.show()
```

```python
print("Uniform Value -> Mean = " , np.mean(uniform_val_data))
print("Normal Value -> Mean = " , np.mean(normal_val_data))
```

```python
print("Uniform Value -> Variance = " , np.var(uniform_val_data))
print("Normal Value -> Variance = " , np.var(normal_val_data))
```

```python
print("Uniform Value -> Standard deviation = " , np.std(uniform_val_data))
print("Normal Value -> Standard deviation = " , np.std(normal_val_data))
```
